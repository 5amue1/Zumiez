```python
import requests
import re
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
URL = 'https://www.zumiez.com/skate/complete-skateboards.html'
page = requests.get(URL)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
count=1
skt_urls=[]
for i in soup.findAll('a', { "class" : "product-name"}):
        skt_urls.append(i.get('href'))

skt_urls
```

```python
df = pd.DataFrame(columns = ['Brand', 'Deck', "Construction", "Length", "Width", "Trucks", "Wheels", "Wheelbase", "Bearings"]) 
```

```python
skate_info = []
for temp in skt_urls:

    temp_page = requests.get(temp)
    temp_soup = BeautifulSoup(temp_page.content, 'html.parser')

    x = [x.text.lower()[:-1] for x in temp_soup.find('td', { "class" : "data"}).find('ul').findAll('li')]
    
    brand = deck = construction = length = width = trucks = wheels = wheelbase = bearings = "default"
    
    for line in x:
        try:
            if "nyjah huston pro" in line:
                break
            elif "skateboard complete" in line:
                if "from" in line and brand=="default":
                    brand = re.findall(r"(?<=from ).*",line)[0]
                elif brand=="default":
                    brand = re.findall(r"(?<=by ).*",line)[0] 
                if deck=="default":
                    deck = re.findall(r".*?(?=[0-9.\"]* skateboard complete)",line)[0]
            elif "construction" in line and construction=="default":
                construction = line.split("construction")[0]
            elif "length" in line and length=="default":
                length = float(line.split(':')[1].strip().split('"')[0])
            elif "width" in line and width=="default":
                width = float(line.split(':')[1].strip().split('"')[0])
            elif "trucks" in line and trucks=="default":
                trucks = line
            elif "wheels" in line and wheels=="default":
                wheels = re.findall(r"[0-9]*mm",line)[0].split("mm")[0]
            elif "wheelbase" in line and wheelbase=="default":
                wheelbase = float(line.split(':')[1].strip().split('"')[0])
            elif "bearings" in line and bearings=="default":
                bearings = line.split("bearings")[0].strip()
        except:
            continue
    
    df.loc[len(df)] = [brand, deck, construction, length, width, trucks, wheels, wheelbase, bearings]
```
